```python
import wordcloud
import matplotlib.pyplot as plt
%matplotlib inline
keywords = {'Hello':10, 'Python':15, 'Crawl':5, 'Visual':10}

wc = wordcloud.WordCloud()
cloud = wc.generate_from_frequencies(keywords)

plt.figure()
plt.imshow(cloud)
plt.show()
```
